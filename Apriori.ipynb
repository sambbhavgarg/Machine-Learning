```python
def createC1(dataset):
    "Create a list of candidate item sets of size one."
    c1 = []
    for transaction in dataset:
        print("-",transaction)
        for item in transaction:
            if not [item] in c1:
                print("--",[item])
                c1.append([item])
    c1.sort()
    #frozenset because it will be a key of a dictionary.
    return list(map(frozenset, c1))#frozen set makes lists unchangable and available for 
                                   #making it a key in a dictionary
```

```python
def scanD(dataset, candidates, min_support):
    "Returns all candidates that meets a minimum support level"
    sscnt = {}
    for tid in dataset:
        print("--",tid)
        for can in candidates:
            print("-",can)
            if can.issubset(tid):#if dataset transac is present in candidates, 
                sscnt.setdefault(can, 0)#set value of frozen set to 0
                sscnt[can] += 1#and increment by 1

    num_items = float(len(dataset))
    retlist = []
    support_data = {}
    for key in sscnt:
        support = sscnt[key] / num_items
        if support >= min_support:
            retlist.insert(0, key)
        support_data[key] = support
    return retlist, support_data
```

```python
def aprioriGen(Lk, k): #creates a list of combination using simple increment in a list and making pairs
    retList = []
    lenLk = len(Lk)
    for i in range(lenLk):
        for j in range(i+1, lenLk): 
            L1 = list(Lk[i])[:k-2]
            #slicing kiya hai to check starting ke first k-2 elements same hain ke nahi
            print("--",L1)
            L2 = list(Lk[j])[:k-2]
            print("--",L2)
            L1.sort()#
            L2.sort()
            if L1==L2: #if first k-2 elements are equal
                retList.append(Lk[i] | Lk[j]) #set union eliminates  duplicates
                print(retList)
        
    return retList
```

```python
def apriori(dataSet, minSupport = 0.40):
    C1 = createC1(dataSet)
    D = list(map(set, dataSet))
    L1, supportData = scanD(D, C1, minSupport)
    L = [L1]
    k = 2
    while (len(L[k-2]) > 0):
        Ck = aprioriGen(L[k-2], k)
        Lk, supK = scanD(D, Ck, minSupport)#scan DB to get Lk
        supportData.update(supK)
        L.append(Lk)
        k += 1
        
       
    return L, supportData
```

```python
def generateRules(L, supportData, minConf=0.7):  #supportData is a dict coming from scanD; 
    bigRuleList = []
    for i in range(1, len(L)):#only get the sets with two or more items; to traverse list
        for freqSet in L[i]:
            print(freqSet)
            H1 = [frozenset([item]) for item in freqSet]
            if (i > 1):
                rulesFromConseq(freqSet, H1, supportData, bigRuleList, minConf)
            else:
                calcConf(freqSet, H1, supportData, bigRuleList, minConf)
    return bigRuleList   
```

```python
def calcConf(freqSet, H, supportData, brl, minConf=0.7):
    prunedH = [] #create new list to return
    for conseq in H:
        conf = supportData[freqSet]/supportData[freqSet-conseq] #calc confidence
        if conf >= minConf: 
            print (freqSet-conseq,'-->',conseq,'conf:',conf)
            brl.append((freqSet-conseq, conseq, conf))
            prunedH.append(conseq)
    return prunedH
```

```python
def rulesFromConseq(freqSet, H, supportData, brl, minConf=0.7):
    m = len(H[0])
    if (len(freqSet) > (m + 1)): #try further merging
        Hmp1 = aprioriGen(H, m+1)#create Hm+1 new candidates
        Hmp1 = calcConf(freqSet, Hmp1, supportData, brl, minConf)
        if (len(Hmp1) > 1):    #need at least two sets to merge
            rulesFromConseq(freqSet, Hmp1, supportData, brl, minConf)
```
